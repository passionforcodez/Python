```python
import os
import io
from IPython.display import Image, display, HTML
from PIL import Image
import base64
from google.colab import userdata

# Access the Hugging Face API key from Colab secrets
hf_api_key = userdata.get('HuggingFace')

# You can add a check here to ensure the key was retrieved successfully
if hf_api_key is None:
  print("HuggingFace API key not found in Colab secrets. Please add it to the Secrets Manager.")
```

```python
# Helper function
import requests, json

#Summarization endpoint
def get_completion(inputs, parameters=None,ENDPOINT_URL="https://api-inference.huggingface.co/models/facebook/bart-large-cnn"):
    headers = {
      "Authorization": f"Bearer {hf_api_key}",
      "Content-Type": "application/json"
    }
    data = { "inputs": inputs }
    if parameters is not None:
        data.update({"parameters": parameters})
    response = requests.request("POST",
                                ENDPOINT_URL, headers=headers,
                                data=json.dumps(data)
                               )
    return json.loads(response.content.decode("utf-8"))
```

```python
text = ('''The tower is 324 metres (1,063 ft) tall, about the same height
        as an 81-storey building, and the tallest structure in Paris.
        Its base is square, measuring 125 metres (410 ft) on each side.
        During its construction, the Eiffel Tower surpassed the Washington
        Monument to become the tallest man-made structure in the world,
        a title it held for 41 years until the Chrysler Building
        in New York City was finished in 1930. It was the first structure
        to reach a height of 300 metres. Due to the addition of a broadcasting
        aerial at the top of the tower in 1957, it is now taller than the
        Chrysler Building by 5.2 metres (17 ft). Excluding transmitters, the
        Eiffel Tower is the second tallest free-standing structure in France
        after the Millau Viaduct.''')

get_completion(text)
```

```python
import gradio as gr

def summarize(input):
    output = get_completion(input)
    return output[0]['summary_text']

gr.close_all()

# Create a larger Textbox for the output with a specified number of lines
output_box = gr.Textbox(label="Output", lines=10)

demo = gr.Interface(fn=summarize, inputs="text",
                    outputs=output_box)
demo.launch(share=True)
```

```python
from transformers import pipeline

# Assuming get_completion is now the NER pipeline from transformers
def get_completion(text):
    # Initialize the NER pipeline with aggregation_strategy='simple'
    ner_pipeline = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")
    return ner_pipeline(text)

# The rest of your code from the image can remain the same
def ner(input_text):
    output_entities = get_completion(input_text)
    # The output from the pipeline will already be in the correct format for Gradio
    return {"text": input_text, "entities": [(entity['word'], entity['entity_group']) for entity in output_entities]}

# Example usage
text = "My name is Andrew, I'm building DeepLearningAI and I live in California"
output = get_completion(text)
print(output)
```
